# NetworkX Tutorial Cambridge

```python
import networkx as nx
```

```python
# Quick example
g = nx.Graph()
g.add_edge('a','b',weight=0.1)
g.add_edge('b','c',weight=1.5)
g.add_edge('a','c',weight=1.0)
g.add_edge('c','d',weight=2.2)
print(nx.shortest_path(g,'b','d'))
print(nx.shortest_path(g,'b','d',weight='weight'))
```

```python
# Begin with empty graph
g = nx.Graph()

# One node at the time
g.add_node(1)

# Add a list of nodes
g.add_nodes_from([2,3])

# Add container of nodes
h = nx.path_graph(5)
g.add_nodes_from(h)

# Remove nodes
g.remove_node(2)

# Check nodes
g.nodes()
```

```python
# Begin with empty graph
g = nx.Graph()

# A node can be a string...
g.add_node('string')

# A node can be a function...
import math 
g.add_node(math.cos)

# A node can be a file...
f = open('temp.txt','w')
g.add_node(f)

# Check nodes
g.nodes()
```

```python
# Begin with empty graph
g = nx.Graph()

# Add a single edge
g.add_edge(1,2)

# Add an unpacked tuple
e=(3,4)
g.add_edge(*e)

# Add a list of edges
g.add_edges_from([(5,6),(7,8)])

# Add edges from container
h = nx.path_graph(5)
g.add_edges_from(h.edges())

# Remove an edge
g.remove_edge(1,2)

# Check graph
print('Nodes:', g.nodes())
print('Edges:', g.edges())
```

```python
# Begin with empty graph
g = nx.Graph()

# Add edges
g.add_edges_from([(1,2),(1,3)])

# Add different node
g.add_node('a')

# Check graph - NODES
print()
print(' Nodes:', g.nodes())
print('---sum:', g.number_of_nodes())
print('-order:', g.order())

# Check graph - EDGES
print()
print(' Edges:', g.edges())
print('---sum:', g.number_of_edges())
print('--size:', g.size())

# Check individual node
print()
print('Neighbors of 1:', list(g.neighbors(1)))
print('   Degree of 1:', g.degree(1))
```

```python
# Begin with empty graph
g = nx.Graph()

# Add node with extra info
g.add_node(1, time='10am')
print()
print(g.node[1])
print(g.node[1]['time'])

# Add edge with weight
g.add_edge(3,2,weight=0.8)
print()
print(g[3][2])
print(g[3][2]['weight'])
```

```python
# Begin with empty graph
g = nx.Graph()

# Node iterators
g.add_edge(1,2)
print()
for node in g.nodes():
    print('node_iteration: ', node, g.degree(node))
    
# Edge iteration
g.add_edge(1,2,weight=1.2)
g.add_edge(1,3,weight=2.0)
print()
for n1, n2 in g.edges():
    print('edge_iteration 1: ', n1, n2)
print()
for n1, n2, attr in g.edges(data=True):
    print('edge_iteration 2: ', n1, n2, attr, attr['weight'])
```

```python
# Begin with an empty directed-graph
dg = nx.DiGraph()

# Add weithed edges
dg.add_weighted_edges_from([(1,4,0.5),(3,1,2.2)])

# Calculate degrees (in, out, total)
print()
print('>Weight Outwards: ', dg.out_degree(1, weight='weight'))
print('>Weight Inwards: ', dg.in_degree(1, weight='weight'))
print('>Weight: ', dg.degree(1, weight='weight'))

# Calculate siccessors and predecessors
print()
print('+Predecessors(1): ', list(dg.predecessors(1)))
print('+Successors(1): ', list(dg.successors(1)))

# Check graph
print()
print('--Nodes:', g.nodes())
print('--Edges:', g.edges())
```

```python
# Import matlab
import matplotlib.pyplot as plt

# Begin with smaple graph
petersen = nx.petersen_graph()
tutte = nx.tutte_graph()
maze = nx.sedgewick_maze_graph()
tet = nx.tetrahedral_graph()
k_5 = nx.complete_graph(5)
k_3_5 = nx.complete_bipartite_graph(3,5)
barbelle = nx.barbell_graph(10,10)
lollipop = nx.lollipop_graph(10,20)
er = nx.erdos_renyi_graph(100,0.15)
ws = nx.watts_strogatz_graph(30,3,0.1)
ba = nx.barabasi_albert_graph(100,5)
red = nx.random_lobster(100,0.9,0.9)

# Iterable with names
grafos = [petersen, tutte, maze, tet, k_5, k_3_5, barbelle, lollipop, er, ws, ba, red]

# Plot
for grafo in grafos:
    print('++++',grafo,'+++')
    nx.draw(grafo)
    plt.show()
```

```python
# Union
k_3 = nx.complete_graph(3)
j_3 = nx.complete_graph(6)
j_3.remove_nodes_from(k_3.nodes())

# Operate
grafo = nx.union(k_3, j_3)

# Plot
nx.draw(grafo)
plt.show()
```

```python
# subgraph
g = nx.complete_graph(6)

# Operate
grafo = nx.subgraph(g,[0,3,4,5])

# Plot
nx.draw(grafo)
plt.show()
```

```python
# cartesian product
g1 = nx.lollipop_graph(3,5)
g2 = nx.Graph()
g2.add_edge(99,100)

# Operate
grafo = nx.cartesian_product(g1,g2)

# Plot
nx.draw(grafo)
plt.show()
```

```python
# compose
g1 = nx.complete_graph(6)
g2 = nx.Graph()
g2.add_edges_from([(5,6),(6,7),(7,5)])


# Operate
grafo = nx.compose(g1,g2)

# Plot
nx.draw(grafo)
plt.show()
```

```python
# complement
g1 = nx.tetrahedral_graph()

# Operate
grafo = nx.complement(g1)

# Plot
nx.draw(grafo)
plt.show()
```

```python
# empty copy
g1 = nx.lollipop_graph(3,5)

# Operate
grafo = nx.create_empty_copy(g1)

# Plot
nx.draw(grafo)
plt.show()
```

```python
# convert to undirected and back
g1 = nx.tetrahedral_graph()

# Operate
grafo = nx.to_directed(g1)

# Plot
nx.draw(grafo)
plt.show()

# Operate
grafo = nx.to_undirected(grafo)

# Plot
nx.draw(grafo)
plt.show()
```
