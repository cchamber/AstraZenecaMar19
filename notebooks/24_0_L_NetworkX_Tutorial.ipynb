# NetworkX Tutorial 1

```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create empty graph
g = nx.Graph()
```

```python
# Add node
g.add_node(2)
g.add_node(5)

# Add endge
g.add_edge(2,5)
```

```python
# Add nodes from existing edge
g.add_edge(4,1)
```

```python
# Add from many to many
g.add_edges_from([(2,5),(1,4)])
```

```python
# Echo info
print(nx.info(g))
```

```python
# Draw
nx.draw(g)
plt.show()
```

```python
# Read edges form file
filename='../src/visualization/links.csv'
h = nx.read_weighted_edgelist(filename, create_using=nx.Graph(), nodetype=str, delimiter=',', comments='#')
print(nx.info(h))
```

```python
nx.draw(h)
plt.show()
```
